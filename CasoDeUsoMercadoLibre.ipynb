### Importación paqueterias necesarias

```python
import requests
import pandas as pd
import regex as re
```

### Obtención de request por país

```python
#Se obtienen codigos de paises

country = requests.get('https://api.mercadolibre.com/sites')
country = country.json()
country = pd.json_normalize(country)
country['request'] = 'https://api.mercadolibre.com/sites/'+country['id']+'/categories'
country


```

### Obtención de categoria de videojuegos para los paises

```python
aux = pd.DataFrame()

for i in range(len(country['request'])):
    categories = requests.get(country['request'][i])
    categories = categories.json()
    categories = pd.json_normalize(categories)
    categories = categories.query('name.str.startswith("Consolas")| name.str.startswith("Games")',engine="python")
    categories = categories.rename(columns = {'id':'id_categoria'})
    categories['id'] = categories['id_categoria'].apply(lambda x : re.sub('[0-9]', '', x))
    temp = pd.merge(country, categories,how='inner', on='id')
    aux = pd.concat([aux,temp])

categories = aux
categories
```

### Obtención de usuarios para México por categoria de videojuegos

```python
#Obtiene id de los vendedores de las categoria consolas y videojuegos

aux = pd.DataFrame()
temp = pd.DataFrame()

for j in range(1,5,1):
    users_mexico = requests.get('https://api.mercadolibre.com/sites/MLM/search?category=MLM1144&offset='+str(j*50))
    users_mexico = users_mexico.json()
    for i in range(50):
        temp = pd.json_normalize(users_mexico['results'][i]['seller'])
        aux = pd.concat([aux,temp])

users_mexico = aux
aux = pd.DataFrame()
temp = pd.DataFrame()

users_mexico = users_mexico.drop_duplicates('id')
#Nickname e ID de los vendedores de mexico para la categoria de Videojuegos
users_mexico


```

### Obtención de reputación de los vendedores obtenidos

```python
#Reputacion del vendedor
aux = pd.DataFrame()
temp = pd.DataFrame()

for i in range(len(users_mexico)):
    users_mexico_reputacion = requests.get('https://api.mercadolibre.com/users/'+str(users_mexico.iat[i, 0]))
    users_mexico_reputacion = users_mexico_reputacion.json()
    df_reputacion = pd.DataFrame({
            'ciudad': users_mexico_reputacion['address']['city'],
            'estado':users_mexico_reputacion['address']['state'],
            'tipo_usuario': users_mexico_reputacion['user_type'],
            'status_vendedor':users_mexico_reputacion['status']['site_status'],
            'tipo_periodo_transacciones':users_mexico_reputacion['seller_reputation']['transactions']['period'],
            'total_transacciones':users_mexico_reputacion['seller_reputation']['transactions']['total'],
            'reputacion_vendedor':users_mexico_reputacion['seller_reputation']['level_id'],
            'reputacion_vendedor_status':users_mexico_reputacion['seller_reputation']['power_seller_status'],
            'id':users_mexico_reputacion['id']}, index = [1])
    aux = pd.concat([aux,df_reputacion])

users_mexico_reputacion = aux
aux = pd.DataFrame()
temp = pd.DataFrame()

users_mexico_reputacion



```

```python
users_mexico_reputacion = pd.merge(users_mexico_reputacion, users_mexico,how='inner', on='id')
users_mexico_reputacion
```

### Escritura de archivo con los resultados de reputación por vendedor

```python
users_mexico_reputacion.to_csv('ReputacionPorVendedor.csv', index=False) 
```
